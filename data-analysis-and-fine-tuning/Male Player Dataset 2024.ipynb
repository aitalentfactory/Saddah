# Each Column Description

**1. Core Player Information**

    Name: The player's full name (e.g., Kylian Mbappé).

    Nation: The player's nationality (e.g., France, Norway).

    Club: The team the player currently plays for (e.g., Paris SG, Manchester City).

    Position: Primary playing position (e.g., ST = Striker, CM = Central Midfielder, GK = Goalkeeper).

    Age: The player's age (in years).

    Overall: Overall skill rating (0–99), reflecting their total ability in the game.

**2. Key Performance Attributes**

    Pace: Speed (0–99), combining acceleration and sprint speed.

    Shooting: Goal-scoring ability (0–99), including accuracy and power.

    Passing: Accuracy and effectiveness of passes (0–99).

    Dribbling: Ball control and agility while moving (0–99).

    Defending: Defensive skills (0–99), including tackling and marking.

    Physicality: Strength and endurance (0–99).

**3. Detailed Physical & Technical Attributes**

    Acceleration: How quickly the player reaches top speed (0–99).

    Sprint: Maximum running speed (0–99).

    Positioning: Offensive awareness (for attackers) or defensive positioning (for defenders) (0–99).

    Finishing: Accuracy in front of goal (0–99).

    Shot (Power): Strength of shots (0–99).

    Long (Shots): Accuracy of long-range shots (0–99).

    Volleys: Ability to score from mid-air shots (0–99).

    Penalties: Success rate in penalty kicks (0–99).

    Vision: Ability to spot and execute passes (0–99).

    Crossing: Accuracy of crosses from wide areas (0–99).

    Free (Kicks): Accuracy in free-kick situations (0–99).

    Curve: Ability to bend the ball’s trajectory (0–99).

    Agility: Quickness in changing direction (0–99).

    Balance: Ability to stay upright during physical challenges (0–99).

**4. Mental & Ball Control Attributes**

    Reactions: Speed of response to in-game events (0–99).

    Ball (Control): Technical control of the ball (0–99).

    Composure: Calmness under pressure (0–99).

**5. Defensive & Aerial Attributes**

    Interceptions: Ability to intercept passes (0–99).

    Heading: Accuracy in aerial duels (0–99).

    Def (Awareness): Positioning and decision-making in defense (0–99).

    Standing (Tackle): Effectiveness of standing tackles (0–99).

    Sliding (Tackle): Effectiveness of sliding tackles (0–99).

    Jumping: Vertical leap height (0–99).

**6. Physical & Work Rate Traits**

    Stamina: Endurance over the course of a match (0–99).

    Strength: Physical power in challenges (0–99).

    Aggression: Tendency to engage in physical duels (0–99).

    Att Work Rate: Effort in attacking phases (Low/Medium/High).

    Def Work Rate: Effort in defensive phases (Low/Medium/High).

**7. Player Preferences**

    Preferred Foot: Dominant foot (Left/Right).

    Weak Foot: Skill level with the non-preferred foot (1–5 stars).

    Skill Moves: Ability to perform flair dribbles (1–5 stars).

**8. Meta Data**

    URL: Link to the player’s EA Sports FC profile.

    Gender: Player’s gender (M = Male in this dataset).

    GK: Goalkeeper-specific rating (0–99). Only relevant for goalkeepers (e.g., Thibaut Courtois has 93 here, while others may have NaN/blank).

# Data Preparation for Fine Tunning Model

```python
import pandas as pd
import json

# Load your dataset
df = pd.read_csv("/content/drive/MyDrive/dataset-competition/male_players.csv")

# Filter out goalkeepers or include them (adjust as needed)
# df = df[df['Position'] != 'GK']

# Define a function to generate prompts/completions
def generate_prompt_completion(row):
    prompt = f"Describe {row['Name']}'s profile, including their club, position, and key attributes."

    completion = (
        f"{row['Name']} is a {row['Age']}-year-old {row['Position']} from {row['Nation']}, "
        f"playing for {row['Club']}. Key attributes: Overall ({row['Overall']}), "
        f"Pace ({row['Pace']}), Shooting ({row['Shooting']}), Dribbling ({row['Dribbling']}), "
        f"Passing ({row['Passing']}), and Physicality ({row['Physicality']})."
    )

    return {"prompt": prompt, "completion": completion}

# Generate JSONL data
jsonl_data = []
for _, row in df.iterrows():
    jsonl_data.append(generate_prompt_completion(row))

# Save to a .jsonl file
with open("/content/drive/MyDrive/dataset-competition/footballers_finetuning.jsonl", "w") as f:
    for entry in jsonl_data:
        f.write(json.dumps(entry) + "\n")
```

```python
def generate_training_program(row):
    """
    Generates prompt-completion pairs for training schedule recommendations.
    """
    # Base template for prompts
    prompt = f"Create a 4-week training program for {row['Name']} ({row['Position']}, aged {row['Age']}, " \
             f"with the following attributes: Pace {row['Pace']}, Shooting {row['Shooting']}, " \
             f"Stamina {row['Stamina']}, Strength {row['Strength']}. Focus on improving weaknesses."

    # Customize program based on position and attributes
    position = row['Position']
    program = []

    # Base training components (ensure all positions have ≥3 items)
    base_components = {
        "ST": ["Finishing drills", "Speed training", "Positioning exercises"],
        "CM": ["Passing patterns", "Long-range shooting", "Tactical awareness"],
        "CB": ["Aerial duels", "Tackling practice", "Defensive positioning"],
        "GK": ["Reaction drills", "Cross claiming", "Penalty area command"],
        "LB": ["Defensive sprints", "Crossing practice", "Overlap drills"],
        "RW": ["Dribbling circuits", "Cut-inside finishing", "Crossing accuracy"],
        "CAM": ["Through-ball drills", "Shooting from distance", "Positional rotation"],
        "CDM": ["Interception drills", "Long passing", "Tackling technique"]
    }

    # Get components or default to general training
    program = base_components.get(position, ["Strength training", "Agility drills", "Tactical sessions"])

    # Add weakness-specific training
    weaknesses = []
    if row['Stamina'] < 70:
        weaknesses.append("endurance training")
    if row['Strength'] < 75:
        weaknesses.append("strength & conditioning")
    if row['Balance'] < 80:
        weaknesses.append("core stability exercises")

    if weaknesses:
        program.append("Special focus on: " + ", ".join(weaknesses))

    # Ensure program has at least 3 items (pad with defaults if needed)
    while len(program) < 3:
        program.append("Recovery sessions")

    # Structure completion with training schedule
    completion = f"""4-week training program for {row['Name']} ({position}):
Week 1-2: Foundation Phase
- Monday: {program[0]} (High intensity)
- Wednesday: {program[1]} (Moderate intensity)
- Friday: {program[2]} (Low intensity)
- Daily: {' & '.join(weaknesses) if weaknesses else 'No specific focus'} sessions

Week 3-4: Intensity Phase
- Add 20% increased load
- Focus on {position}-specific game simulations
- Recovery sessions every 3rd day"""

    return {"prompt": prompt, "completion": completion}
```

```python
def prepare_training_dataset(df, output_file="/content/drive/MyDrive/dataset-competition/training_programs.jsonl"):
    """
    Converts footballer dataset to training recommendation format.
    """
    jsonl_data = []

    for _, row in df.iterrows():
        try:
            # Generate program for all positions (including GK)
            entry = generate_training_program(row)
            jsonl_data.append(entry)
        except Exception as e:
            print(f"Skipping row {_}: {e}")
            continue

    # Check if data was generated
    if not jsonl_data:
        raise ValueError("No training data generated. Check your input data and filters.")

    # Save to JSONL file
    with open(output_file, "w") as f:
        for entry in jsonl_data:
            f.write(json.dumps(entry) + "\n")

    print(f"Successfully generated {len(jsonl_data)} entries in {output_file}")
    return jsonl_data
```
